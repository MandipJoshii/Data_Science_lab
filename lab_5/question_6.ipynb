### 6. Identify outliers in the Marks column using the IQR method.
### Remove the outliers from the dataset.

```python
import pandas as pd

df = pd.read_csv('../Book1.csv')

print("original data: ")
print(df.head())
```

### INTER QUARTILE RANGE

```python
q1 = df['score'].quantile(0.25)
q3 = df['score'].quantile(0.75)

iqr = q3 - q1

lower_limit = q1 - 1.5 * iqr
upper_limit = q3 + 1.5 * iqr

outliers = df[(df['score'] < lower_limit) | (df['score'] > upper_limit)]
print("\nOutliers:")
print(outliers)

print(f"IRQ IS {iqr:.2f} OF LOWER LIMIT:{lower_limit:.2f} AND UPPER LIMIT: {upper_limit:.2f} ")
print(df[(df['score'] >= lower_limit) & (df['score'] <= upper_limit)])
```
